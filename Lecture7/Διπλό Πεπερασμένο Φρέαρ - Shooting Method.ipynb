# Επίλυση διπλού πεπερασμένου τετραγωνικού φρέατος

Η  χρονοανεξάρτητη εξίσωση Schrodinger δίνεται από τη σχέση:
$$ E\Psi(x)=\left( \frac{-\hbar^2}{2m} \nabla^2 + V(x)\right) \Psi(x) $$
Θα λύσουμε αυτήν την εξίσωση αριθμητικά για να αποκτήσουμε τις ενεργειακές ιδιοκαταστάσεις.  
Σε μία διάσταση, μπορούμε να αναδιατάξουμε την εξίσωση οδηγώντας στην:
$$\psi''(x) =  -\frac{2m}{\hbar^2}\left[ E - V(x)\right]\psi(x)$$
Οπου το διπλό τετραγωνικό φρέαρ δίνεται από την:
$$V(x)=-V_0,\; \frac{b}{2}<|x|<\frac{b}{2}+a$$
$$V(x)=0,\; εκτός$$
Θα ορίσουμε $ m =\hbar = 1 $, έτσι ώστε να πρέπει να λύσουμε τελικώς την:
$$\psi''(x) =  2\left[ V(x)-E\right]\psi(x)$$ 

#### Τρόπος επίλυσης
Πρέπει να εργαστούμε με διακριτά χωρικά πλέγματα.  Δεδομένου ότι το δυναμικό είναι συμμετρικό, οι ιδιοσυναρτήσεις θα είναι είτε συμμετρικές είτε αντι-συμμετρικές, (Βλ. Zettili παρ. 4.7.2 για b=0 και εξίσωση 4.4) οπότε πρέπει να λύσουμε μόνο για το "μισό" της κυματομορφής, από $x = 0$ έως πχ. $ x = L $.  Η λύση της κυματομορφής θα είναι σε ένα διακριτό πλέγμα (πεδίο ορισμού) με $ N $ σημεία, που ορίζονται από τις:
$$ x_i = i\times\delta x$$
$$\delta x = \frac{L}{N}$$
Επιπλέον, πρέπει να χρησιμοποιήσουμε πεπερασμένη διαφορά για να υπολογίσουμε τις παραγώγους αριθμητικά στο διακριτό πλέγμα.  Για παράδειγμα, το εμπρός σύστημα πεπερασμένων διαφορών για την πρώτη παράγωγο είναι
$$\psi'(x_i) \approx \frac{\psi(x_{i+1})- \psi(x_i)}{\delta x}$$
Ενώ το πίσω είναι:
$$\psi'(x_i) \approx \frac{\psi(x_{i})- \psi(x_{i-1})}{\delta x}$$
Χρησιμοποιώντας τόσο τα συστήματα διαφορισης προς τα εμπρός όσο και προς τα πίσω, μπορούμε να αποκτήσουμε αυτό που ονομάζεται Δεύτερη Προσέγγιση Κεντρικής Διαφοράς στο Δεύτερο Παράγωγο:
$$\psi''_i \approx (\psi'_i)'=\left( \frac{\psi_{i+1} - \psi_i}{\delta x} \right)'=\frac{(\psi_{i+1}-\psi_i) - (\psi_i - \psi_{i-1})}{\delta x^2}$$

$$\psi''_i \approx \frac{\psi_{i+1} - 2\psi_i + \psi_{i-1}}{\delta x ^2}$$
όπου $\psi_i=\psi(x_i)$.

#### Shooting method
Ας εφαρμόσουμε την τελευταία έκφραση για τη διακριτή παράγωγο δεύτερης τάξης στην εξίσωση Schrodinger. Αναδιατάσσοντας, παίρνουμε
$$ \psi_{i+1} = 2\psi_i - \psi_{i-1} + 2\delta x^2 \left[ V(x_i) -E \right] \psi_i$$
Οι λύσεις είναι είτε συμμετρικές είτε αντι-συμμετρικές, οπότε είτε το $ \psi_0 $ ή το $ \psi_0 '$ θα είναι μηδενικά. Από την 
$$\psi_1 \approx \psi_0 + \delta x \psi_0'$$ (Ανάπτυγμα Taylor πρώτης τάξης)
Μπορούμε να ξεκινήσουμε είτε με $ \psi_0 $ ή $ \psi_0 '$, και στη συνέχεια να χρησιμοποιήσουμε τους προηγούμενους δύο τύπους για να κατασκευάσουμε το υπόλοιπο της κυματομορφής.  Αυτή η διαδικασία ονομάζεται shooting method.

Σημείωση: Πώς γνωρίζουμε τα $\psi_0 $ ή $ \psi_0 '$; Αποδεικνύεται οτι δεν έχει σημασία. Υποθέτουμε απλώς μια τιμή και το μόνο αποτέλεσμα που έχει η τιμή είναι στην κανονικοποίηση της προκύπτουσας κυματοσυνάρτησης, την οποία μπορούμε εύκολα να διορθώσουμε.

Από τη φυσική μας διαίσθηση (που αποκτήθηκε από το πρόβλημα 2.47 στο Griffiths), η βασική κατάσταση πρέπει να είναι συμμετρική, με $ \psi_0 '= 0 $, ενώ η πρώτη διεγερμένη κατάσταση θα είναι αντι-συμμετρική, με $ \psi_0 = 0 $.




#### Υπολογίζοντας την ενέργεια
Η ενέργεια τώρα εμφανίζεται ως παράμετρος της shooting method. Επιπλέον, αν και μπορούμε να λύσουμε την εξίσωση Schodinger για οποιοδήποτε $ E $, μόνο πολύ συγκεκριμένες ενέργειες θα δώσουν φυσικές, κανονικοποιημένες καταστάσεις.

Πως τις βρίσκουμε: ξεκινάμε με ένα ενεργειακό εύρος, $E\in \left[E_{min}, E_{max} \right]$ για το οποιο θεωρούμε ότι περιέχει τις φυσικές καταστάσεις. Ελέγχουμε αυτήν την περιοχή με μια μέθοδο του τύπου [bisection](https://en.wikipedia.org/wiki/Bisection_method) μέχρι να βρούμε την «πραγματική» ενέργεια σε έναν αυθαίρετο βαθμό ακρίβειας. 

Με αυτή τη μέθοδο, δεν θα έχουμε ποτέ τη σωστή ενέργεια, πράγμα που σημαίνει ότι η κυματομορφή θα είναι ακόμα λίγο μακριά και θα "εκτοξευθεί" σε μεγάλο $ x $.  Αλλά, θα είμαστε μ'αυτό τον τρόπο(αυθαίρετα) κοντά στη σωστή ενέργεια έτσι ώστε η κυματομορφή του διπλού φρέατος  να είναι πολύ κοντά στη σωστή. Έτσι, μόλις φτάσουμε στην επιθυμητή ακρίβεια, απλώς περιορίζουμε την κυματομορφή, μηδενίζοντας  τις τιμές πέρα από μια ορισμένη απόσταση από το κέντρο των πηγαδιών, κανονικοποιώντας οτι απομένει.

```python
%matplotlib ipympl
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.widgets import Slider, Button, RadioButtons
import numpy
import numpy.polynomial.hermite as Herm
import math
```

Συναρτήσεις που θα χρησιμοποιηθούν:

1. *Το δυναμικό* (Potential) παίρνει μια array από τα σημεία του πλέγματος και επιστρέφει μια array του ίδιου μήκους με τις τιμές του δυναμικού σε κάθε σημείο, $V_(x_i) = V_i $.
2. *H norm* παίρνει μια σειρά από τιμές για την κυματοσυνάρτηση $ \psi(x_i) = \psi_i $ και επιστρέφει τη norm, $\int \psi^2dx \approx \sum_i \psi_i^2 \delta x$.
3. H *normalize* θα κανονικοποιήσει την κυματοσυνάρτηση έτσι ώστε $\sum_i \psi_i^2 \delta x=1$.

```python
def Potential(x, a, b, V0):
    V=numpy.zeros(len(x))
    V[(abs(x)>=b/2.) & (abs(x)<b/2.+a)] = -V0
    return V

def norm(dx, psi):
    return dx*numpy.dot(psi,psi)

def normalize(dx, psi):
    return psi/numpy.sqrt(norm(dx,psi))
```

Η βάση τώρα του προγράμματος.  Ξεκινώντας από την τιμή της κυματομορφής και την παράγωγό της στο $ x = 0 $, χρησιμοποιούμε τη shooting method για να υπολογίσουμε το υπόλοιπο της κυματομορφής επανελημμένως, αλλάζοντας την ενέργεια μέχρι να βρούμε μια αποδεκτή λύση.

```python
def forward_shooting(V, x, dx, f0, df0, Emin, Emax, acc):
    """
    V - array με τις τιμές του δυναμικού (βλ. Potential function)
    x - τα στοιχεία του δικτύου του πεδίου ορισμού
    dx - Η απόσταση των στοιχείων του δικτύου του πεδίου ορισμού
    f0 - Τιμή της συνάρτησης στο x=0
    df0 - Η τιμή της παραγώγου της συνάρτησης στο x=0
    Emin, Emax - Περιοχή τιμών για έλεγχο λύσεων
    acc - Περιοχή ακρίβειας λύσης 
    """
    Edelta = 1.
    E = Emin
    psi = numpy.zeros(len(x))
    psi[0]=f0
    max_steps = 500
    count=0
    
    # Βρίσκουμε την ενέργεια που αντιστοιχεί σε μια φυσική κατάσταση μέσω διχοτόμησης (bisection)
    # Επαναλαμβάνουμε μέχρι να έχουμε την ενέργεια μέσα στο acc ή μέχρι να πάρουμε τις επαναλήψεις Max_Steps
    while Edelta > acc and count < max_steps:
        count+=1
        # Περνώντας από τα σημεία του πλέγματος βρίσκουμε την τιμή της κυματοσυνάρτησης από την shooting method
        for i, x_ in enumerate(x[0:-1]):
            if i==0:
                psi[i+1] = psi[i] + dx*df0
            else:
                psi[i+1] = -psi[i-1] + 2.*( 1 + dx**2 * (V[i] - E) )*psi[i]
                
            # Αυτό είναι το δύσκολο κομμάτι! Εάν η  κυματομορφή εκτιναχθεί αριθμητικά,
            # Χρησιμοποιούμε τη διχοτόμηση για να αλλάξουμε την ενέργεια και να αναζητήσουμε τη φυσική ενέργεια.
            # Οι λεπτομέρειες για το πώς να επιλέξουμε την επόμενη ενέργεια εξαρτάται από τον αριθμό των κόμβων που έχει η κυματοσυνάρτηση.
            # Αυτό θα λειτουργήσει για n = 0,1,4,5,8,9 ...
            if(psi[i] > 25):
                Emin=E
                E = Emin + (Emax-Emin)/2.
                break
            elif(psi[i] < -25):
                Emax=E
                E=Emax-(Emax-Emin)/2.
                break
        Edelta = (Emax-Emin)
    return (psi, E)
    
```

Η επόμενη συνάρτηση δένει τα πάντα: Όταν την καλέσουμε με τις παραμέτρους διπλού φρεατίου, παίρνουμε τη βασική και τις πρώτες ενεργειακές καταστάσεις, καθώς και τις ενέργειές τους.

```python
def acquire_double_well_states(a, b, V0, x, dx, acc, L_plot):
    V = Potential(x, a, b, V0)
    
    # Βασική κατάσταση (συμμετρική, psi '(x = 0) = 0)
    (psi_0, energy_0) = forward_shooting(V, x, dx, 1, 0, -V0, 0, acc)
    
    # Πρώτη διεγερμένη κατάσταση (αντισυμμετρική, PSI (x = 0) = 0)
    (psi_1, energy_1) = forward_shooting(V, x, dx, 0, 1, -V0, 0, acc)
    psi_0[x>L_plot]=0
    psi_1[x>L_plot]=0

    # Χρησιμοποιηούμε τη συμμετρία για να αποκτήσουμε το άλλο μισό της κυματοσυνάρτησης
    space = numpy.zeros(2*len(x))
    space[0:len(x)]=-x[::-1]
    space[len(x):2*len(x)]=x

    Psi_0 = numpy.zeros(len(space))
    Psi_0[0:len(x)]=psi_0[::-1]
    Psi_0[len(x):2*len(x)]=psi_0
    Psi_0=normalize(dx, Psi_0)

    Psi_1 = numpy.zeros(len(space))
    Psi_1[0:len(x)]=psi_1[::-1]
    Psi_1[len(x):2*len(x)]=-psi_1
    Psi_1=normalize(dx, Psi_1)
    
    return (space, Psi_0, energy_0, Psi_1, energy_1)
    
```

Ας ελέγξουμε τις λύσεις για ένα συγκεκριμένο $ b $.  Παρατηρήστε ότι και οι δυο καταστάσεις είναι σωστά κανονικοποιημένες.

```python
a=1.5
b=0
V0=2
dx=0.05
acc=1e-10
L=10*(b+a)
L_plot=10
x = numpy.linspace(0, L, int(L/dx) + 1, True)

(space, Psi_0, energy_0, Psi_1, energy_1) = acquire_double_well_states(a, b, V0, x, dx, acc, L_plot)

print("Norm Psi_0=",norm(dx, Psi_0))
print("Energy_0=", energy_0)
print("Norm Psi_1=",norm(dx, Psi_1))
print("Energy_1=", energy_1)

plt.subplots()
plt.plot(space, Potential(space, a, b, V0), label="V(x)")

plt.plot(space, Psi_0, 'r',label="$\psi_0$")
plt.plot([-L_plot,L_plot], [energy_0,energy_0],'r--',label="$E_0$")
plt.plot(space, Psi_1, 'g',label="$\psi_1$")
plt.plot([-L_plot,L_plot], [energy_1,energy_1],'g--',label="$E_1$")

plt.xlim([-L_plot,L_plot])
plt.ylim([-2,0.6])
plt.legend(loc=3)
plt.title('Ground and First Excited States for the Double Well')
plt.xlabel('x')
plt.show()
```

Τώρα ας δημιουργήσουμε ένα animated γράφημα, όπου μπορούμε να αλλάξουμε τον διαχωρισμό του φρεατίου, $ b $, και να δούμε πώς αλλάζουν καταστάσεις και οι ενέργειες.

```python
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.15, bottom=0.15)

# Ρυθμίστε μια αρχική λύση για να σχεδιάσετε
b0=0
(space, Psi_0, energy_0, Psi_1, energy_1) = acquire_double_well_states(a, b0, V0, x, dx, acc, L_plot)

pot, = plt.plot(space, Potential(space, a, b0, V0), label="$V(x)$")
state_0, = plt.plot(space, Psi_0, 'r',label="$\psi_0$")
E0, = plt.plot([-L_plot,L_plot], [energy_0,energy_0],'r--',label="$E_0$")
state_1, = plt.plot(space, Psi_1, 'g',label="$\psi_1$")
E1, = plt.plot([-L_plot,L_plot], [energy_1,energy_1],'g--',label="$E_1$")

plt.xlim([-L_plot,L_plot])
plt.ylim([-2,0.7])
plt.legend(loc=3)
plt.title('Ground and First Excited States for the Double Well')
plt.xlabel('x')

# SLider για να αλλάξoυμε διαδραστικά τον διαχωρισμό φρεατίου, b
axb = plt.axes([0.25, 0.04, 0.65, 0.03])
sb = Slider(axb, 'b', 0., 4., valinit = b0)

# Όταν μετακινείται ο slider, υπολογίζονται οι λύσεις για το νέο b
def update(val):
    bnew = sb.val
    (space, Psi_0, energy_0, Psi_1, energy_1) = acquire_double_well_states(a, bnew, V0, x, dx, acc, L_plot)
    pot.set_ydata(Potential(space, a, bnew, V0))
    state_0.set_ydata(Psi_0)
    E0.set_ydata([energy_0,energy_0])
    state_1.set_ydata(Psi_1)
    E1.set_ydata([energy_1,energy_1])
    
sb.on_changed(update)

plt.show()
```
